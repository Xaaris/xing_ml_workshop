Simple gradient descent examples.

## Notation

When a function $f(x)$ has a single parameter the derivative is written as: 
$$
\frac{d}{dx}f(x) = \frac{d f(x)}{dx}
$$

You can think of $d$ as an operator that maps a function of one parameter to the derivative of the function.

When a function $f(x)$ has multiple parameter we compute derivatives with respect to one variable at a time, these are called **partial derivatives**.

Instead of using operator $\frac{d}{dx}$, the partial derivative operator is ∂ (a stylized d), called differential operator. For a function $f(x,y)$ the partial derivatives are $\frac{∂}{∂x}f(x,y)$ and $\frac{∂}{∂y}f(x,y)$, they are often just called partials. The partial derivative with respect to $x$ is just the usual scalar derivative, simply treating any other variables in the equation as a constant.

The vector of partial derivatives is called the **gradient** of $f(x)$:
$$
\nabla f(x,y) = [\frac{∂f(x,y)}{∂x}, \frac{∂f(x,y)}{∂y}]
$$

When working with neuronal networks we usually have a loss function $L(x;\theta)$ that has parameter $x$ and $\theta$. For optimization we only want to calculate the gradient w.r.t $\theta$. For this case we put a subscript to the nabla operator: $\nabla_\theta f(x,y)$.

## Gradient descent

Let $y(x)$ be a function that returns a scalar and $x$ is a vector. The goal is to minimize $y$, this means to find the value $x$ where $y(x)$ has the minimal value.

Using gradient descent we apply the following update rule:

$$
x_{k+1} = x_k + \alpha \nabla_x y(x_k)
$$


In every iteration we want to update $\theta$ so that the function moves a little bit closer to its minimum. In every iteration we have to make the following decisions for each parameter in $\theta$:
 * direction of update
 * magnitude of update

## Batch gradient descent

**Batch gradient optimization** referred to the case where the gradient computations were
accumulated over one iteration over the entire training set before being applied to the parameters,
**Stochastic gradient methods** were typically online methods with parameter update based on a
single training example.

Current deep learning stochastic gradient algorithms instead use parameter updates based on gradient averages over small subsets (mini-batches) of the full training set, and the term *batch size* is commonly used to refer to the size of a mini-batch.

The mini-batch SGD algorithm uses the following update rule:

$$
\theta_{k+1} = \theta_k - \alpha \frac{1}{m} \displaystyle\sum_{x \in B} \nabla_\theta L(x, \theta_k)
$$

where $L$ is the loss function, $B$ is the current batch, $\alpha$ is the learning rate and $m$ is the batch size.

The goal is to use large minibatches in place of small minibatches while maintaining training and evaluation
accuracy. This is of particular interest in distributed learning, because it can allow us to scale to multiple workers by reducing network communication.

**Linear Scaling Rule:** When the minibatch size is multiplied by k, multiply the learning rate by k. See [https://arxiv.org/pdf/1706.02677.pdf](https://arxiv.org/pdf/1706.02677.pdf) for more details.

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

# Single variable example

```python

# Define a function
def y(x):
    return x**2 + 3 * x

# Derive the gradient
def grad_y(x):
    return 2 * x + 3 

# minimum is at: 
# 2x + 3 = 0
# x = -1.5
# y(-1.5) = -2.25

t1 = np.arange(-4.5, 1.5, 0.01)
plt.title('y(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(t1, y(t1))
plt.show()
```

```python
max_steps = 200
alpha = 0.05
initial_x = 5.0 # would be initialized random

x = initial_x
y_hist = []
for i in range(max_steps):
    x = x - alpha * grad_y(x)
    y_hist.append(y(x))
print('Final x=%f' % x)
print('Final y(x)=%f' % y(x))

plt.title('Convergence plot')
plt.xlabel('iteration')
plt.ylabel('y(x)')
plt.plot(x_hist)
plt.show()
```

# Multi variable example

```python

# Function to optimize
def y(x):
    return x[0]**2 + x[1]**2

# Gradient of the function
def grad_y(x):
    return np.array([2 * x[0], 2 * x[1]])
```

```python
max_steps = 300
alpha = 0.01
initial_x = np.array([5.0, 5.0])

x = initial_x
y_hist = []
for i in range(max_steps):
    x = x - alpha * grad_y(x)
    y_hist.append(y(x))
print('Final x=%s' % x)
print('Final y(x)=%s' % y(x))

plt.title('Convergence plot')
plt.xlabel('iteration')
plt.ylabel('y(x)')
plt.plot(x_hist)
plt.show()
```

# TensorFlow example

```python
import numpy as np
import tensorflow as tf
```

```python

max_steps = 200
alpha = 0.01

# y = x_0^2 + x_1^2
x = tf.Variable(np.array([5.0, 3.0]), name="x")
y = tf.reduce_sum(tf.pow(x, 2))

optimizer = tf.train.GradientDescentOptimizer(alpha)
train_op = optimizer.minimize(y)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    y_hist = []
    for i in range(max_steps):
        x_val, y_val, _ = sess.run([x, y, train_op])
        y_hist.append(y_val)
    
print('Final x=%s' % x_val)
print('Final y(x)=%s' % y_val)
    
plt.title('Convergence plot')
plt.xlabel('iteration')
plt.ylabel('y(x)')
plt.plot(y_hist)
plt.show()    
    
```

##  Gradient calculation

```python
# y = x_0^2 + x_1^2
# dy_dx_0 = 2x_0
# dy_dx_1 = 2x_1
x = tf.Variable(np.array([5.0, 3.0]), name="x")
y = tf.reduce_sum(tf.pow(x, 2))
dy_dx = tf.gradients(y, x)

with tf.Session() as sess:
    dy_dx_val = sess.run(dy_dx, {x: [1.0, 4.0]})[0]
    print('dy/dx_0 = %f' % dy_dx_val[0])
    print('dy/dx_1 = %f' % dy_dx_val[1])
    
```
