## A short timeline of Machine Learning

### 1950s-1960s
Pioneering machine learning research is conducted using simple algorithms. First neuronal networks. Machines play checkers and tic-tac-toe.

### 1970s
'AI' winter caused by pessimism about machine learning effectiveness.

## 1980s
ML shifts from a knowledge-driven approach to a data-driven approach. Scientists begin creating programs to analyze large amounts of data and learn from the results.

### 1986
Invention of Automatic Differentation (Backpropagation)

### 1992
TD-Gammon, a computer backgammon program that uses an artificial neural network

### 1997

IBM's Deep Blue beats the world champion Kasparov at chess

### 2010 
Kaggle, a website that serves as a platform for machine learning competitions, is launched.

Large Scale Visual Recognition Challenge (ILSVRC) introduced the **Imagenet Challenge**. Imagenet is a dataset of 1.2 million images from 1000 categories. The goal is to predict the image category.

### 2012

AlexNet, a deep convolutional neural network architecture won the Imagenet challenge with a top-5 error of 15.3%. 

The Imagenet 2012 event was probably what triggered the big explosion of AI today.

### 2013
Deepmind releases a reinforcement learning algorithm that learns to play Atari games from pixel input only.

### 2015
**AlphaGo** wins 5-0 in the game of Go against 3-times European Champion Fan Hui. The results were published in full technical detail in the Nature journal.

### 2016
**AlphaGo** wins 4-1 against legendary player Mr Lee Sedol, winner of 18 world titles and widely considered to be the greatest player of the past decade.

Google announces availability of the **tensor processing unit** (TPU), a chip developed by Google specifically for neural network machine learning.

### 2017

29 out of 38 teams that took part in the Imagenet challenge have a top-5 error of 5% or better.

DeepMind released **AlphaZero**, which, after 24 hours training, wins against the best programs in Go, chess and shogi. AlphaZero was trained solely by "self-play".
