## Setup environment

It is recommended to work with the [Anaconda](https://www.anaconda.com/download) Python distribution.

See [Gist](https://gist.github.com/asprenger/e26d5ce4f8dbde51f1ec95238b68c304) for details.

### Install Python libraries

The following Python libraries are required:

* Keras
* Tensorflow
* Gensim

Install Python libraries:

    pip install keras==2.2.2
    pip install tensorflow==1.9.0
    pip install gensim==3.6.0

### Downlad workshop material

Clone Github repository

    git clone https://github.com/asprenger/xing_ml_workshop.git

Start jupyter:

    cd xing_ml_workshop
    jupyter notebook

This should open up a browser window.

## Notes about Jupyter Notebooks

Notebooks show up with ending `.ipynb`, they are opened by clicking on the link in the Jupyter web UI. 

Each notebooks runs in its own OS process.

A notebook contains cells that are of type `code` or `markdown`. 

`code` cells are executing Python code. All `code` cells in a notebook share a variable context

The content of `markdown` cells is rendered in Markdown format.

A cell can be edited by double clicking it and is executed by pressing `ctrl-enter`.

A notebook can be restarted from the Kernel item in the menue bar.

Notebooks can be saved by pressing `cmd-s`. When a notebook is saved all output is saved as well, including plots generated by Python. This is awesome for sharing results and ideas. 

For more details on Jupyter Notebooks check out the [documentation](http://jupyter.org/documentation).
