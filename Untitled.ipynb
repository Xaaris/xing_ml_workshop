$f(x) = w_0 x_0^2 + w_1 x_1^2 + b $

$\frac{\partial f(x)}{\partial x_0} = 2 w_0 x_0$

$\frac{\partial f(x)}{\partial x_1} = 2 w_1 x_1$

$\frac{\partial f(x)}{\partial b} = 1$

$loss(f(x), y) = (f(x) - y)^2 = f(x)^2 - 2 f(x) y + y^2$

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, w, b):
    assert x.shape[0] == 2
    assert w.shape[0] == 2
    assert b.shape[0] == 1
    return w[0] * x[0]**2 + w[1] * x[1]**2 + b[0]

true_w = np.array([0.8, 1.3])
true_b = np.array([0.5])

# initialize parameters
#w = np.array([0.01, -0.02])
#b = np.array([0.2])

w = np.random.normal(loc=0.0, scale=0.1, size=(2,))
b = np.random.normal(loc=0.0, scale=0.1, size=(1,))

alpha = 0.0001

losses = []
for i in range(200):
    
    x = np.random.rand(2) * 5.0
    y = f(x, true_w, true_b)

    y_pred = f(x, w, b)
    
    loss = (y_pred - y)**2
    losses.append(loss)

    dloss_dw0 = 2 * (y_pred - y) * x[0]**2
    dloss_dw1 = 2 * (y_pred - y) * x[1]**2
    dloss_db = 2 * (y_pred - y)
    
    w = w - alpha * np.array([dloss_dw0, dloss_dw1])
    b = b - alpha * np.array([dloss_db])

print(w, b)    
plt.figure(figsize=(10,10))
plt.plot(losses)
plt.show()
```
